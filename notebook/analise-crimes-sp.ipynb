#### Reduzindo o dataset para trabalhar somente com o que queremos (Previsão de crimes)

```python
df_reduced = df[['DATA_OCORRENCIA_BO', 'HORA_OCORRENCIA_BO', 'BAIRRO', 'LATITUDE', 'LONGITUDE', 'RUBRICA']]
```

```python
df_reduced.head()
```

#### Tratamento de nulos + criação de colunas derivadas

```python
df_reduced.isnull().sum()
```

```python
df_reduced = df_reduced.loc[
    df_reduced['BAIRRO'].apply(lambda x: isinstance(x, str)) &  # Garante que é string
    df_reduced['BAIRRO'].str.match(r'^[A-Za-zÀ-ÿ\s]+$', na=False)  # Só letras, espaços e hífen
]
```

```python
df_reduced = df_reduced.loc[df_reduced['HORA_OCORRENCIA_BO'].notna()]
```

```python
df_reduced.loc[:, 'BAIRRO'] = df_reduced['BAIRRO'].fillna('Não informado')
```

```python
df_reduced = df_reduced.loc[df_reduced['LATITUDE'].notna() & df_reduced['LONGITUDE'].notna()]
```

```python
df_reduced.isnull().sum()
```

```python
df_reduced.head()
```

```python
df_reduced['HORA_OCORRENCIA_BO'] = pd.to_datetime(df_reduced['HORA_OCORRENCIA_BO'], format='%H:%M:%S', errors='coerce')
```

```python
df_reduced.loc[:, 'HORA'] = df_reduced['HORA_OCORRENCIA_BO'].dt.hour
df_reduced.loc[:, 'MINUTO'] = df_reduced['HORA_OCORRENCIA_BO'].dt.minute
```

#### Modelagem 

```python
# Passo 1: pegar todos os valores únicos
bairros = df_reduced['BAIRRO'].unique()
dias_semana = range(7)  # 0 a 6
horas = range(24)       # 0 a 23

# Passo 2: criar todas as combinações possíveis
combinacoes = pd.DataFrame(product(bairros, dias_semana, horas), columns=['BAIRRO', 'DIA_SEMANA', 'HORA'])

# Passo 3: contar ocorrências reais
ocorrencias_reais = df_reduced.groupby(['BAIRRO', 'DIA_SEMANA', 'HORA']).size().reset_index(name='TOTAL')

# Passo 4: juntar com as combinações
crimes_por_bairro_hora = combinacoes.merge(ocorrencias_reais, on=['BAIRRO', 'DIA_SEMANA', 'HORA'], how='left')

# Passo 5: preencher onde não teve crime com 0
crimes_por_bairro_hora['TOTAL'] = crimes_por_bairro_hora['TOTAL'].fillna(0).astype(int)

# Passo 6: variável alvo: 1 se teve crime, 0 se não teve
crimes_por_bairro_hora['OCORRENCIA'] = (crimes_por_bairro_hora['TOTAL'] > 0).astype(int)
```

```python
crimes_por_bairro_hora.head()
```

```python
# Features e alvo
X = crimes_por_bairro_hora[['BAIRRO', 'DIA_SEMANA', 'HORA']]
y = crimes_por_bairro_hora['OCORRENCIA']
```

```python
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42
)
```

```python
# Definir o pré-processador
preprocessor = ColumnTransformer(
    transformers=[
        ('bairro', OneHotEncoder(handle_unknown='ignore'), ['BAIRRO'])  # codifica bairros
    ],
    remainder='passthrough'  # mantém as outras colunas (dia da semana, hora)
)
```

```python
# Definir o pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, max_depth=100, class_weight='balanced', random_state=42))
])
```

```python
# Treinar o modelo
pipeline.fit(X_train, y_train)
```

```python
# Fazer previsões
y_pred = pipeline.predict(X_test)
```

```python
# Dados de entrada: Bairro, Dia da semana (sexta-feira = 5), Hora (22h)
# Vamos usar um DataFrame para essa entrada
input_data = pd.DataFrame({
    'BAIRRO': ['IPIRANGA'],
    'DIA_SEMANA': [5],  # Sexta-feira
    'HORA': [9]  # 22h
})

# Fazer a previsão com o pipeline
predicao = pipeline.predict(input_data)

# Exibir a previsão
print(predicao)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Gera e imprime a matriz de confusão
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()

# Mostra o relatório com métricas
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred))
```

```python
importances = pipeline.named_steps['classifier'].feature_importances_
nomes_features = pipeline.named_steps['preprocessor'].get_feature_names_out()

# Criar dataframe com as features e suas importâncias
df_importance = pd.DataFrame({'feature': nomes_features, 'importance': importances})

# Ordenar pelas mais importantes
df_importance = df_importance.sort_values(by='importance', ascending=False)

# Mostrar top 10
plt.figure(figsize=(10, 6))
sns.barplot(data=df_importance.head(10), x='importance', y='feature')
plt.title("Top 10 Features mais importantes para o modelo")
plt.xlabel("Importância")
plt.ylabel("Variável")
plt.tight_layout()
plt.show()
```
